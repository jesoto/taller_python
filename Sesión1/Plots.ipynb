```python
px.scatter(data,'gdpPercap', 'lifeExp',color="country")
```

```python
medianGapminder = data.groupby(['continent','year']).lifeExp.median()
medianGapminder
medianGapmindergdp = data.groupby(['continent','year']).gdpPercap.median()
medianGapmindergdp
```

```python
data.lifeExp.mean() #el promedio de expectativa de vida
data.gdpPercap.mean()
```

Si queremos ver que continente esta debajo de la expectativa de vida promedio, podemos gregar esta linea al gráfico y visualizar mejor

```python
plt.figure(figsize=(12,6))
# Now plot it
plt.axhline(y = data.lifeExp.mean(), linestyle = ':', label = "Expectativa de vida promedio")
sns.lineplot(x="year",y="lifeExp",hue="continent",data= medianGapminder.reset_index())
plt.title("Expectativa dee vida por continente 1952-2007")
```

En este grafico tengo este insight

```python
plt.figure(figsize=(12,6))
# Now plot it
plt.axhline(y = data.gdpPercap.mean(), linestyle = ':', label = "PBI per Capita promedio")
sns.lineplot(x="year",y="gdpPercap",hue="continent",data= medianGapmindergdp.reset_index())
plt.title("PBI per Capita promedio por continente 1952-2007")
```

```python
data['gdp'] = data['pop'] * data['gdpPercap']
```

```python
mediangdp = data.groupby(['continent','year']).gdp.mean()
```

```python
plt.figure(figsize=(12,6))
# Now plot it
plt.axhline(y = data.gdp.mean(), linestyle = ':', label = "PBI per Capita promedio")
sns.lineplot(x="year",y="gdp",hue="continent",data= mediangdp.reset_index())
plt.title("PBI per Capita promedio por continente 1952-2007")
```

```python
# Box plot of GDP per capita stratified by continent. Remember to set the figure size before you call the plot in order for figsize to work. 

plt.figure(figsize=(12, 6))
sns.boxplot(data.continent, data.gdpPercap, palette = "Spectral")
plt.title("Box Plot of GDP per Capita by Continent")
plt.xlabel("Continent")
plt.ylabel("GDP per Capita [in USD]")
```
