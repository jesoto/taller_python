# Dataframes

```python
import pandas as pd
import numpy as np
import os
```

### Inicializar un dataframe con diccionarios

```python
dicc_a = {
        'id': ['1', '2', '3', '4', '5'],
        'nombre': ['Nombre1', 'Nombre2', 'Nombre3', 'Nombre4', 'Nombre5'], 
        'apellido': ['Ape1', 'Ape2', 'Ape3', 'Ape4', 'Ape5']}

#recordar que un diccionario se puede convertir en un dataframe como lo he hecho aquí.
df_from_dict = pd.DataFrame(dicc_a, columns = ['id', 'nombre', 'apellido'])
df_from_dict
```

### Inicializar un dataframe con array como filas

```python
my_array = np.array([['Jon',25,1995,2016],['Maria',47,1973,2000],['Bill',38,1982,2005]], dtype=object)

df_from_array = pd.DataFrame(my_array, columns = ['Name','Age','Birth Year','Graduation Year'])
df_from_array
```

### Inicializar un dataframe con arrays como columnas

```python
# List1
Name = ['tom', 'krish', 'nick', 'juli']
  
# List2
Age = [25, 30, 26, 22]
  
# get the list of tuples from two lists.
# and merge them by using zip().
list_of_tuples = list(zip(Name, Age))

df_from_tuples = pd.DataFrame(list_of_tuples, columns = ['Name', 'Age'])
df_from_tuples
```

## Concatenar

```python
raw_data_b = {
        'id': ['4', '5', '6', '7', '8'],
        'nombre': ['Nombre4', 'Nombre5', 'Nombre6', 'Nombre7', 'Nombre8'], 
        'apellido': ['Ape4', 'Ape5', 'Ape6', 'Ape7', 'Ape8']}
df_b = pd.DataFrame(raw_data_b, columns = ['id', 'nombre', 'apellido'])
df_b
```

```python
df_nueva = pd.concat([df_a, df_b], axis=0) 
df_nueva
```

```python
df_nueva.reset_index(drop=True,inplace=True)
df_nueva
```

### Concatenar por columnas

```python
pd.concat([df_a, df_b], axis=1)
```

```python
raw_data_c = {
        'id': ['4', '5', '6', '7', '8'],
        'edad': [20, 25, 30, 22, 40]}
df_c = pd.DataFrame(raw_data_c, columns = ['id', 'edad'])
df_c
```

```python
df_nuevo = pd.concat([df_b.set_index('id'), df_c.set_index('id')['edad']], axis=1)
df_nuevo.reset_index(inplace=True, drop=False)
df_nuevo
```

## Merge - Equivalente a Joins en SQL

### Argumento ON

```python
pd.merge(df_b, df_c, on=['id'])
```

```python
pd.merge(df_b, df_c, left_on=['id'], right_on=['id'])
```

### Argumento how

```python
df_a
```

```python
df_b
```

```python
#Outer Join: La union de todo lo que existe, tambien conocido como full join

pd.merge(df_a, df_b, on='id', how='outer')
```

```python
#Inner Join: La union en la interseccion de ambos conjuntos 

pd.merge(df_a, df_b, on='id', how='inner')
```

```python
#Right Join: La union tomando de pivot el universo de la derecha

pd.merge(df_a, df_b, on='id', how='right')
```

```python
#Right Join: La union tomando de pivot el universo de la izquierda

pd.merge(df_a, df_b, on='id', how='left')
```

### Argumento suffixes

```python
#Cuando las columnas posean el mismo nombre, al aplicar pd.merge() por defecto Pandas agregará al nombre un x e y. 
#Con suffixes se puede agregar un nombre a elección:

pd.merge(df_a, df_b, on='id', how='left', suffixes=('_data1', '_data2'))
```

```python
column_replace = []

for (col,metric) in df_dep.columns:
        new_col = col + "_" + metric
        column_replace.append(new_col)
```

```python
df_dep.columns = column_replace
```

```python
df_dep.reset_index(inplace=True,drop=False)
```
